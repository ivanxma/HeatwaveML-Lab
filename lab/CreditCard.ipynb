```python
%%sql

# delete the training model for the log anomaly
delete from ML_SCHEMA_admin.MODEL_CATALOG where train_table_name = 'CreditCardAnomalyDB.credit_card_train';
select * from ML_SCHEMA_admin.MODEL_CATALOG;


SET @semi_supervised_model='anomaly_detection_semi_supervised_use_case';
CALL sys.ML_TRAIN('CreditCardAnomalyDB.credit_card_train', "target", 
                          CAST('{"task": "anomaly_detection", "experimental": {"semisupervised": {}}}' as JSON), 
                          @semi_supervised_model);
```

```python
%%sql
SELECT model_id, model_handle, train_table_name FROM ML_SCHEMA_admin.MODEL_CATALOG  
        WHERE model_handle = 'anomaly_detection_semi_supervised_use_case';
```

```python
%%sql
CALL sys.ML_MODEL_LOAD(@semi_supervised_model, NULL);
```

```python
%%sql

CALL sys.ML_PREDICT_TABLE('CreditCardAnomalyDB.credit_card_test', @semi_supervised_model, 
                                  'CreditCardAnomalyDB.credit_card_predictions_semi', JSON_OBJECT('threshold', 0.55));
```
