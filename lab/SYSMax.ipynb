```python
%%sql
set @modelid = 'sysmax.model02';
```

```python
%%sql
call sys.ML_TRAIN("sysmax.mytable2", null, JSON_OBJECT('task', 'anomaly_detection', 'contamination', 0.01, 
                                                      'exclude_column_list', JSON_ARRAY('my_row_id')), @modelid);
```

```python
%%sql
SELECT JSON_EXTRACT(model_metadata, '$.contamination') 
          FROM ML_SCHEMA_admin.MODEL_CATALOG 
          WHERE model_handle = @modelid;
```

```python
%%sql
create table mytest like mytable2;
insert into mytest select mytable2.* from mytable2 left join mytrain on (mytable2.my_row_id = mytrain.my_row_id) where mytrain.my_row_id is null limit 10;
```

```python
%%sql
select * from mytest;
```

```python
%%sql
call sys.ML_MODEL_LOAD(@modelid, null);
call sys.ML_PREDICT_TABLE('sysmax.mytest', @modelid, 'sysmax.myprediction', JSON_OBJECT('threshold', 0.6));
```
